```python
import numpy as np
import matplotlib.pyplot as plt

# 定义要绘制的数学方程
def equation(x):
    return 1/(1+np.exp(-x))

# 定义 x 的取值范围
x = np.linspace(-100, 100, 10000)

# 计算方程的值
y = equation(x)

# 绘制图形
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Mathematical Equation')
plt.grid(True)

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义要绘制的数学方程
def equation(x,a,b,c,d):
    return a/(b+c*np.exp(-x*d))

# 定义 x 的取值范围
x = np.linspace(-10, 10, 10000)

# 计算方程的值
y = equation(x,a=1,b=1,c=1,d=1)

# 绘制图形
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Mathematical Equation')
plt.grid(True)

# 显示图形
plt.show()
```
