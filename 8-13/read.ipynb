```python
import pandas as pd

# 读取CSV文件
data = pd.read_csv('D:\\Users\\liuyu\\Desktop\\1Fizz_GUI\\8-13\\pupil_positions.csv')

# 显示数据
print(data)
```

```python
selected_columns = ['pupil_timestamp', 'diameter_3d']  # 用你要选择的实际列名替换这些值
data = data[selected_columns]

print(data)
```

```python
#data=data.dropna(axis=0,how='any') #删除含有缺失值的行
print(data)
```

```python
fixed_value=data['pupil_timestamp'].iloc[0]
print(fixed_value)

data['pupil_timestamp']=data['pupil_timestamp']-fixed_value

print(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16, 9),dpi=300)

X=np.array(data['pupil_timestamp'].values.tolist())
Y=np.array(data['diameter_3d'].values.tolist())


#plt.axis('equal')
plt.scatter(X, Y, color='blue',s=1)

plt.xlabel('time')
plt.ylabel('diameter_3d')

plt.grid(True)
plt.show()
```
